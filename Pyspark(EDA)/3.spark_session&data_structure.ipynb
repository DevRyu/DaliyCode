```
%python

from pyspark.sql import SparkSession
# 세션 생성
spark = SparkSession \
  .builder \
  .appName("Spark EDA example on us aircraft data ") \
  .getOrCreate()

#마운트 되어 저장된 csv데이터 읽어서 데이터 프레임 객체에 저장
raw_df = spark.read.csv("/mnt/us-carrier-dataset", header = 'True', inferSchema='True')
```

```
print(type(raw_df))
읽어 드린 데이터 프레임 객체에서의 데이터 형태 확인하는 메서드
# DataFrame.printSchema: schema 형태를 표시합니다
#       Schema : 어떤 구조로 데이터에 저장되는가를 나타내는 데이터베이스 구조
# DataFrame.count: 데이터 개수를 표시합니다
# DataFrame.describe: 주어진 column 값의 분포 상태(갯수, 평균, 표준편차, 최소, 최대)를 포함하는 Dataframe을 리턴합니다
# 여기에 다시 show 메소드를 호출해 주면 해당 값들이 표시됩니다. column을 지정해 주지 않으면 모든 column에 대해서 분포 상태를 구합니다

# DataFrame.show: 주어진 개수만큼의 데이터를 표시합니다. 지정하지 않을 경우 20개를 표시합니다
```
